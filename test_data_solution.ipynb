```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_test = pd.read_csv("test.csv")
```

```python
df_test.head()
```

```python
df_test["PoolQC"] = df_test["PoolQC"].fillna(0)
df_test["ExterQual"] = df_test["ExterQual"].fillna(0)
df_test["ExterCond"] = df_test["ExterCond"].fillna(0)
df_test["BsmtQual"] = df_test["BsmtQual"].fillna(0)
df_test["BsmtCond"] = df_test["BsmtCond"].fillna(0)
df_test["HeatingQC"] = df_test["HeatingQC"].fillna(0)
df_test["KitchenQual"] = df_test["KitchenQual"].fillna(0)
df_test["FireplaceQu"] = df_test["FireplaceQu"].fillna(0)
df_test["GarageQual"] = df_test["GarageQual"].fillna(0)
df_test["GarageCond"] = df_test["GarageCond"].fillna(0)
df_test['LotFrontage'] = df_test['LotFrontage'].fillna(df_test['LotFrontage'].mean())
```

```python
df_test = df_test.replace("Ex", 5)
df_test = df_test.replace("Gd", 4)
df_test = df_test.replace("TA", 3)
df_test = df_test.replace("Fa", 2)
df_test = df_test.replace("Po", 1)
```

```python
df_test["CentralAir"] = df_test["CentralAir"].replace("Y", 1)
df_test["CentralAir"] = df_test["CentralAir"].replace("N", 0)
#df["CentralAir"].value_counts() #YES or No variable
```

```python
df_test["BsmtExposure"] = df_test["BsmtExposure"].fillna(0)
df_test["BsmtExposure"] = df_test["BsmtExposure"].replace(4, 3)
df_test["BsmtExposure"] = df_test["BsmtExposure"].replace("Av", 2)
df_test["BsmtExposure"] = df_test["BsmtExposure"].replace("No", 0)
df_test["BsmtExposure"] = df_test["BsmtExposure"].replace("Mn", 1)
#df["BsmtExposure"].value_counts()
```

```python
df_test["PavedDrive"] = df_test["PavedDrive"].replace("Y", 2)
df_test["PavedDrive"] = df_test["PavedDrive"].replace("N", 0)
df_test["PavedDrive"] = df_test["PavedDrive"].replace("P", 1)
#df["PavedDrive"].value_counts()
```

```python
#df_test.drop("Id", axis=1, inplace = True) #first of all we drop because it was assigned randomly and brings no value
df_test.drop("GarageCars", axis=1, inplace = True) # we drop GarageCars because it is logical that number of cars depends on the garage area
df_test.drop("PoolQC", axis=1, inplace = True)
df_test.drop("MiscFeature", axis = 1, inplace = True)
df_test.drop("BedroomAbvGr", axis = 1, inplace = True)
df_test.drop("1stFlrSF", axis = 1, inplace = True)
df_test.drop("2ndFlrSF", axis = 1, inplace = True)
df_test.drop("GarageCond", axis = 1, inplace = True)
df_test.drop("LowQualFinSF", axis = 1, inplace = True)
df_test.drop("KitchenAbvGr", axis = 1, inplace = True) #included in TotalRoomsAboveGround
df_test.drop("Utilities", axis = 1, inplace = True)
df_test.drop('Street', axis = 1, inplace = True)
```

```python
homedata_onehot = pd.get_dummies(df_test, columns=['Foundation', 'GarageFinish', 'LandSlope', 'LotConfig',
 'LandContour', 'GarageType', 'LotShape', 'Alley', 'Fence', 'MSZoning', 'SaleType', 
    'Condition1', 'Functional', 'BsmtFinType1', 'BsmtFinType2',
 'MasVnrType', 'Exterior2nd', 'Heating', 'Neighborhood', 'SaleCondition',
 'Electrical', 'Exterior1st', 'RoofMatl', 'RoofStyle', 'HouseStyle', 'BldgType',
 'Condition2'])
```

```python
homedata_onehot.head()
```

```python
NA = pd.DataFrame(columns=['Column',
                  'perc_NA'])
for column in homedata_onehot.columns:
    NA = NA.append({'Column' : column, 'perc_NA' : sum(homedata_onehot[column].isna())/len(homedata_onehot)}, ignore_index=True)
```

```python
NA.sort_values('perc_NA', ascending=False).head(11)
```

```python
same_count=0
diff_count=0
total_diff = 0
for i, x in zip(homedata_onehot['YearBuilt'],homedata_onehot['GarageYrBlt']):
    if np.isnan(x):
        continue
    else:
        if i == x:
            same_count+=1
        else:
            diff_count+=1
            total_diff= total_diff+(x-i)
        
print('same count:', same_count)
print('diff_count:', diff_count)
print('avg diff:', total_diff/diff_count)
print(same_count/(same_count+diff_count))
print(1-(same_count/(same_count+diff_count)))
print(81*0.79)
print(81*0.21)
```

```python
homedata_onehot['GarageYrBlt'] = homedata_onehot['YearBuilt']
```

```python
print(sum(homedata_onehot['MasVnrArea'].isna()))
homedata_onehot['MasVnrArea'] = homedata_onehot['MasVnrArea'].fillna(0)
```

```python
NA = pd.DataFrame(columns=['Column',
                  'perc_NA'])
for column in homedata_onehot.columns:
    NA = NA.append({'Column' : column, 'perc_NA' : sum(homedata_onehot[column].isna())/len(homedata_onehot)}, ignore_index=True)
NA.sort_values('perc_NA', ascending=False)
```

```python
homedata_onehot = homedata_onehot.fillna(0)
```
